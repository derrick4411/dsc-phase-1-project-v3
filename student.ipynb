## DATA CLEANING


### looking for any duplicate
this code helps to identify any column or rows duplicated

```python
# looking for duplicates
df.duplicated().sum()
```

```python
#removing the duplicates
df.drop_duplicates(inplace=True)
df
```

### missing values
this code helps to identify various values missing in the data set

```python
#missing values
df.isnull().sum()
```

```python
# handle missing values for specific categorical columns
categorical_cols = ['Weather.Condition', 'Broad.phase.of.flight', 'Purpose.of.flight',
                   'Aircraft.Category', 'Engine.Type']
df[categorical_cols] = df[categorical_cols].fillna('Unknown')
df.head()
```

```python
#filling null value in report status
df['Report.Status'].fillna('no_report', inplace=True)
#last 7 rows
df.tail(7)
```

```python
#drop null values
df.dropna(inplace=True)
```

## Generating Graphs
This markdown cell indicates the start of the graphing section.

### Bar Chart of Total fatal accident
This code cell generates a bar chart visualizing the total accident inlocation aganist  total fatal injuries.

```python
# Bar chart of Total fatal accident
plt.figure(figsize=(10, 6))
plt.bar(df['Engine.Type'].astype(str), df['Total.Fatal.Injuries'], width=0.7, edgecolor="white", linewidth=0.7)
plt.title("Total number of death")
plt.xlabel('Engine.Type')
plt.ylabel('Total.Fatal.Injuries')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Observation
the type of engine tends be crusial as those of uknown shows a greater risk and high possiblity of  causing an accident

**Observations:** The bar chart shows the total fatal injuries in respective to engine model the unkown type tend to cause more accidents


## HISTOGRAM OF an occarence of accident and few get unharmed

frequency of an accident happening and people  get unharmed

```python
# Histogram of unhamed
plt.figure(figsize=(10, 6))
plt.hist(df['Total.Uninjured'].dropna(), bins=5, color="skyblue", edgecolor="black")
plt.title('unharmed')
plt.xlabel('Total.Uninjured')
plt.ylabel('Frequency')
plt.show()
```

## Observation
total number of people who tend to survive and be un harmed tend to be very minimal

## LINE GRAPH OF TOP 5 COUNTRIES AGANIST TOTAL INJURIES
will show occurrence of accident in  a country


```python
#line graph of accident in a country
top_5_countries = df['Country'].value_counts().head(5)
plt.figure(figsize=(10, 6))
plt.plot(top_5_countries)
plt.title('Top 5 Countries with Most Accidents')
plt.xlabel('Country')
plt.ylabel("total_injuries")
plt.show()
```
